## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#changing the data type of date columns
date_cols = [
    'PQ First Sent to Client Date',
    'PO Sent to Vendor Date',
    'Scheduled Delivery Date',
    'Delivered to Client Date',
    'Delivery Recorded Date'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], dayfirst=True, errors='coerce')
````

````python
#checking if data type of date columns changed
df.info()
````

````python
#Changing data type of weight column to numeric
df['Weight (Kilograms)'] = pd.to_numeric(
    df['Weight (Kilograms)']
    .astype(str)
    .str.replace(r'[^0-9.]', '', regex=True),
    errors='coerce'
)
````

````python
df.loc[df['Weight (Kilograms)'] > 10000, 'Weight (Kilograms)'] = np.nan
````

````python
df.loc[df['Weight (Kilograms)'] < 0, 'Weight (Kilograms)'] = np.nan
````

````python
df['Weight (Kilograms)'].describe()
````

````python
#Changing the data type of Line Item Insurance Column to Numeric
df['Line Item Insurance (USD)'] = pd.to_numeric(
    df['Line Item Insurance (USD)']
    .astype(str)
    .str.replace(r'[^0-9.]', '', regex=True),
    errors='coerce'
)
````

````python
#Creating delivery delay column to understand delays in delivery
df['Delivery Delay'] = (
    df['Delivered to Client Date'] - df['Scheduled Delivery Date']
).dt.days
````

````python
df['Delivery Delay']
````

````python
# Creating shipment time column to see the duration of shipment
df['Shipment Time'] = (
    df['Delivered to Client Date'] - df['PO Sent to Vendor Date']
).dt.days
````

````python

df['Processing Time'] = (
    df['PO Sent to Vendor Date'] - df['PQ First Sent to Client Date']
).dt.days
````

````python
#Trying to know the orders that are delivered on time
df['On Time'] = df['Delivery Delay'].apply(lambda x: 1 if x <= 0 else 0)
````

````python
df[['Delivery Delay', 'Shipment Time']].describe()
````

### What all manipulations have you done and insights you found?

Since the dataset had inconsistent date formats, I used pd.to_datetime with dayfirst=True and error handling to ensure proper conversion while managing invalid values. I created features like delivery delay, shipment time, and processing time using date columns. I also created an on-time delivery flag to classify shipments as delayed or not. I found inconsistencies such as negative shipment times and extreme delay values, which I treated as data quality issues and handled through filtering.
